```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define the number of classes
num_classes = 10

# Load the pre-trained model from keras.applications with imagenet weights
base_model = keras.applications.ResNet50(
    weights='imagenet',
    input_shape=(224, 224, 3),
    include_top=False
)

# Freeze the base model to prevent its weights from being updated during training
base_model.trainable = False

# Add a new classification layer on top of the pre-trained model
inputs = keras.Input(shape=(224, 224, 3))
x = keras.applications.resnet50.preprocess_input(inputs)
x = base_model(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
outputs = layers.Dense(num_classes, activation='softmax')(x)
model = keras.Model(inputs, outputs)

# Compile the model with a loss function + optimizer
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Load and preprocess the training, validation, and test datasets using ImageDataGenerator
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
train_generator = train_datagen.flow_from_directory(
    './BigCatsKaggle/train/',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)
valid_datagen = ImageDataGenerator(rescale=1./255)
valid_generator = valid_datagen.flow_from_directory(
    './BigCatsKaggle/valid/',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    './BigCatsKaggle/test/',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

# Train the model on the training dataset and validate on the validation dataset
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=valid_generator
)

# Evaluate the model on the test dataset
test_loss, test_acc = model.evaluate(test_generator, verbose=2)
print('Test accuracy:', test_acc)
```
